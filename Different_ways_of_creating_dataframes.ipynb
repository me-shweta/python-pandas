```python
import pandas as pd
#create dataframe using read_csv() method
df = pd.read_csv('/content/weather.csv')
df
```

```python
#create dataframe using python dictionary DataFrame() method
weather_data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temperature': [32,35,28],
    'windspeed': [6, 7, 2],
    'event': ['Rain', 'Sunny', 'Snow']
}
df = pd.DataFrame(weather_data)
df
```

```python
#create dataframe using tuple list DataFrame() method
weather_data = [
    ('1/1/2017', 32, 6, 'Rain'),
    ('1/2/2017', 35, 7, 'Sunny'),
    ('1/3/2017', 27, 2, 'Snow')]
df = pd.DataFrame(weather_data, columns = ["day", "temperature", "windspeed", "event"])
df
```

```python
#Create dataframe using the list of dictionary DataFrame method()
weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
]
df = pd.DataFrame(weather_data)
df
```
