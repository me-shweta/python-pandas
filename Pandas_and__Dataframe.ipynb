###Conditional selection

```python
#dataframe supports sql
#give me all the rows from datframe where my temperature >= 32
df[df.temperature >= 32]
```

```python
#to get a row where temp is maximum
df[df.temperature == df.temperature.max()]
#OR df[df.temperature == df['temperature'].max()] #use this syntax if column name contain spaces
```

```python
#to get a column where temp is maximum
df['day'][df.temperature == df.temperature.max()]
```
